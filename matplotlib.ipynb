```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#adding title and label
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-label")
plt.show()
```

```python
#chalnge line aesthestics
plt.plot(x,y,color="black",linestyle=":",linewidth="3")
plt.title("Dot Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-label")
plt.show()
```

```python
x
```

```python
y1=2*x
y2=3*x
y1
y2
```

```python
#Adding three lines in same plot
plt.plot(x,y1,color="r",linestyle=":",linewidth="3")
plt.plot(x,y2,color="b",linestyle=":",linewidth="3")
plt.title("Two Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-label")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="r",linestyle=":",linewidth="3")
plt.title("Sub plot")
plt.xlabel("X-Label")
plt.ylabel("Y-label")



plt.subplot(2,1,2)
plt.plot(x,y2,color="b",linestyle=":",linewidth="3")
plt.title("Sub Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-label")
plt.grid(True)
plt.show()
```

```python
import numpy as np
student={"Bob":87,"Matt":56,"Anne":100,"Sam":27}
names=list(student.keys())
names
```

```python
values=list(student.values())
values
```

```python
plt.bar(names,values)
plt.show()
```

```python
#Adding title and labels
plt.bar(names,values)
plt.title("Students Marks Distribution")
plt.xlabel("Students Name")
plt.ylabel("Students Marks")
plt.show()
```

```python
#Horizontal Bar Graphs
plt.barh(names,values,color="g")
plt.title("Students Marks Distribution")
plt.xlabel("Students Name")
plt.ylabel("Students Marks")
plt.show()
```

```python
#Adding two plots 
x=[10,20,30,40,50,60,70,80,90]
a=[8,5,4,6,8,9,2,3,8]
b=[7,8,5,6,3,5,62,8,5]
```

```python
plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=100)
plt.show()
```

```python
#Sub SLot
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)


plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```
